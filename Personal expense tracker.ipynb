```python
l1=[]
total_monthly_budget=0
```

```python
def add_expense():
    date= input("When was the expense made? Please enter in the format YYYY-MM-DD ")
    category= input("What was the category of the expense? ")
    price_input= input("How much did you spend? ")
    price= int(price_input) if price_input.strip() else 0
    desc= input("Describe what it was cos you'll forget later :) ")
    d1={"date":date, "category":category, "price":price, "desc":desc}
    l1.append(d1)
    print(l1)
```

```python
add_expense()
```

```python
def view_expense():
    no_of_expenses= len(l1)
    for i in range(no_of_expenses):
        has_blank_entry= False
        print(f"---- Expense {i+1} ----")
        if (l1[i]['date']=='' or l1[i]['category']==''or l1[i]['price']=='' or l1[i]['desc']==''):
            print("ATTENTION!!! FOLLOWING ENTRY IS INCOMPLETE")
        print(f"Date: {l1[i]['date']}")
        print(f"Category: {l1[i]['category']}")
        print(f"Price: {l1[i]['price']}")
        print(f"Description: {l1[i]['desc']}")
        
```

```python
view_expense()
```

```python
def monthly_budget():
    total_monthly_budget= int(input("Enter the monthly budget: "))
```

```python
def total_expenses():
    global total_monthly_budget
    if total_monthly_budget==0:
       total_monthly_budget= int(input("Enter total monthly budget"))
    sum=0
    for i in range(len(l1)):
        sum+= int(l1[i]['price'])
    if sum< total_monthly_budget:
        print(f"Remaining budget for the month is: {total_monthly_budget -sum}")
    else:
              print(f"You have exceeded the monthly budget by: {sum - total_monthly_budget}")
            
        
```

```python
monthly_budget()
```

```python
total_expenses()
```

```python
import csv

def save_expense():
    
    fieldnames = ["date", "category", "price", "desc"]

    csv_filename = "expense_output.csv"

    # Open the CSV file in write mode ('w') with newline='' to prevent extra blank rows
    with open(csv_filename, mode='w', newline='') as file:
        # Create a DictWriter object
        writer = csv.DictWriter(file, fieldnames=fieldnames)

        # Write the header row
        writer.writeheader()

        # Write each dictionary as a row in the CSV file
        writer.writerows(l1)

    print(f"Data successfully written to {csv_filename}")
```

```python
def load_expenses():
    filepath= input("Please enter the filepath/filename of the file you want to load expense from: ")
    with open(filepath, mode='r', newline='', encoding='utf-8') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            l1.append(row)
    return l1
    
```

```python
load_expenses('expense_output.csv')
```

```python
def menu():
    while True:
        print('Press the number options:')
        print('1- Add expense')
        print('2- View expense')
        print('3- Track expense')
        print('4- Save expense')
        print('5- Exit')
        print('6- Enter monthly budget')
        print('7- Load expense')

        choice = input("Enter your choice (1-6): ").strip()

        if choice == "1":
            add_expense()
        elif choice == "2":
            view_expense()
        elif choice == "3":
            total_expenses()
        elif choice == "4":
            save_expense()
        elif choice == "5":
            break
        elif choice== "6":
            monthly_budget()
        elif choice== "7":
            load_expenses()
            
```
